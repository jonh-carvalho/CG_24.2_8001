### Importando bibliotecas


```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Função para plotar pontos
def plot_points(points, color='blue', label=None):
    xs, ys = zip(*points)  # Separa as coordenadas x e y
    plt.scatter(xs, ys, color=color)
    if label:
        for (x, y) in points:
            plt.text(x, y, f'({x}, {y})', fontsize=12, ha='right')
```

# Exercício 1: Translação Simples

```python
# Ponto original
P = np.array([2, 3])
# Vetor de translação
translacao = np.array([4, -2])
# Ponto após a translação
P_prime = P + translacao

plt.figure(figsize=(10, 6))
plt.title('Exercício 1: Translação Simples')
plot_points([P, P_prime], color='blue', label='P e P\'')
plt.text(P[0], P[1], 'P(2, 3)', fontsize=12, ha='right')
plt.text(P_prime[0], P_prime[1], f"P'({P_prime[0]}, {P_prime[1]})", fontsize=12, ha='right')
plt.xlim(0, 10)
plt.ylim(-1, 10)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

# Exercício 2: Escala Uniforme

```python
# Vértices do triângulo original
A = np.array([1, 1])
B = np.array([3, 1])
C = np.array([2, 4])
# Fator de escala
factor_escala = 2
# Novas coordenadas
A_prime = A * factor_escala
B_prime = B * factor_escala
C_prime = C * factor_escala

plt.figure(figsize=(10, 6))
plt.title('Exercício 2: Escala Uniforme')
plot_points([A, B, C], color='blue', label='Triângulo Original')
plot_points([A_prime, B_prime, C_prime], color='red', label='Triângulo Escalado')
plt.text(A[0], A[1], 'A(1, 1)', fontsize=12, ha='right')
plt.text(B[0], B[1], 'B(3, 1)', fontsize=12, ha='right')
plt.text(C[0], C[1], 'C(2, 4)', fontsize=12, ha='right')
plt.text(A_prime[0], A_prime[1], f"A'({A_prime[0]}, {A_prime[1]})", fontsize=12, ha='right', color='red')
plt.text(B_prime[0], B_prime[1], f"B'({B_prime[0]}, {B_prime[1]})", fontsize=12, ha='right', color='red')
plt.text(C_prime[0], C_prime[1], f"C'({C_prime[0]}, {C_prime[1]})", fontsize=12, ha='right', color='red')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

# Exercício 3: Escala Não Uniforme

```python
# Fatores de escala não uniforme
fator_escala_x = 2
fator_escala_y = 0.5
# Aplicando a escala não uniforme
A_prime = np.array([A[0] * fator_escala_x, A[1] * fator_escala_y])
B_prime = np.array([B[0] * fator_escala_x, B[1] * fator_escala_y])
C_prime = np.array([C[0] * fator_escala_x, C[1] * fator_escala_y])

plt.figure(figsize=(10, 6))
plt.title('Exercício 3: Escala Não Uniforme')
plot_points([A, B, C], color='blue', label='Triângulo Original')
plot_points([A_prime, B_prime, C_prime], color='red', label='Triângulo Escalado')
plt.text(A[0], A[1], 'A(1, 1)', fontsize=12, ha='right')
plt.text(B[0], B[1], 'B(3, 1)', fontsize=12, ha='right')
plt.text(C[0], C[1], 'C(2, 4)', fontsize=12, ha='right')
plt.text(A_prime[0], A_prime[1], f"A'({A_prime[0]}, {A_prime[1]})", fontsize=12, ha='right', color='red')
plt.text(B_prime[0], B_prime[1], f"B'({B_prime[0]}, {B_prime[1]})", fontsize=12, ha='right', color='red')
plt.text(C_prime[0], C_prime[1], f"C'({C_prime[0]}, {C_prime[1]})", fontsize=12, ha='right', color='red')
plt.xlim(0, 10)
plt.ylim(-2, 5)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

# Exercício 4: Rotação em Torno da Origem

```python
# Ponto original
P = np.array([1, 0])
# Ângulo de rotação em radianos
theta = np.radians(90)
# Matriz de rotação
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                             [np.sin(theta), np.cos(theta)]])
# Ponto após a rotação
P_prime = rotation_matrix @ P

plt.figure(figsize=(10, 6))
plt.title('Exercício 4: Rotação em Torno da Origem')
plot_points([P, P_prime], color='blue', label='P e P\'')
plt.text(P[0], P[1], 'P(1, 0)', fontsize=12, ha='right')
plt.text(P_prime[0], P_prime[1], f"P'({P_prime[0]:.2f}, {P_prime[1]:.2f})", fontsize=12, ha='right')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

# Exercício 5: Rotação de um Polígono

```python
def rotate_point(x, y, angle):
    # Converte o ângulo de graus para radianos
    theta = math.radians(angle)
    # Aplica a fórmula de rotação
    x_new = x * math.cos(theta) + y * math.sin(theta)
    y_new = -y * math.cos(theta) + x * math.sin(theta)
    return (x_new, y_new)

# Vértices do quadrado
vertices_square = [(1, 1), (1, 4), (4, 4), (4, 1)]
rotated_square = [rotate_point(x, y, -45) for x, y in vertices_square]

plt.figure(figsize=(10, 6))
plt.title('Exercício 5: Rotação de um Polígono')
plot_points(vertices_square, color='red', label='Quadrado Original')
plot_points(rotated_square, color='green', label='Quadrado Rotacionado')
plt.plot(*zip(*vertices_square, vertices_square[0]), color='red', alpha=0.5)
plt.plot(*zip(*rotated_square, rotated_square[0]), color='green', alpha=0.5)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.show()
```

# Exercício 6: Reflexão Simples

```python
# Exercício 6: Reflexão Simples
def reflect_y(x, y):
    return (-x, y)

P_reflection = (2, 5)
reflected_P = reflect_y(*P_reflection)

# Plotando o Exercício 6
plt.figure(figsize=(10, 6))
plt.title('Exercício 6: Reflexão Simples')
plot_points([P_reflection, reflected_P], color='blue', label='P e P\'')
plt.text(P_reflection[0], P_reflection[1], f'P{P_reflection}', fontsize=12, ha='right')
plt.text(reflected_P[0], reflected_P[1], f"P'({reflected_P[0]}, {reflected_P[1]})", fontsize=12, ha='right')
plt.xlim(0, 5)
plt.ylim(0, 6)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.show()
```

# Exercício 7: Reflexão de um Triângulo

```python
# Exercício 7: Reflexão de um Triângulo
def reflect_x(x, y):
    return (x, -y)

triangle_vertices_reflection = np.array([[2, 3], [4, 3], [3, 5]])
reflected_triangle = np.array([reflect_x(x, y) for x, y in triangle_vertices_reflection])

# Plotando o Exercício 7
plt.figure(figsize=(10, 6))
plt.title('Exercício 7: Reflexão de um Triângulo')
plot_points(triangle_vertices_reflection, color='red', label='Triângulo Original')
plot_points(reflected_triangle, color='purple', label='Triângulo Refletido')
plt.plot(*zip(*triangle_vertices_reflection, triangle_vertices_reflection[0]), color='red', alpha=0.5)
plt.plot(*zip(*reflected_triangle, reflected_triangle[0]), color='purple', alpha=0.5)
plt.xlim(0, 5)
plt.ylim(-6, 6)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.show()
```

# Exercício 8: Cisalhamento Horizontal

```python
# Exercício 8: Cisalhamento Horizontal
def shear_horizontal(x, y, k):
    return (x + k * y, y)

P_shear = (2, 3)
sheared_P = shear_horizontal(*P_shear, k=2)

# Plotando o Exercício 8
plt.figure(figsize=(10, 6))
plt.title('Exercício 8: Cisalhamento Horizontal')
plot_points([P_shear, sheared_P], color='blue', label='P e P\'')
plt.text(P_shear[0], P_shear[1], f'P{P_shear}', fontsize=12, ha='right')
plt.text(sheared_P[0], sheared_P[1], f"P'({sheared_P[0]}, {sheared_P[1]})", fontsize=12, ha='right')
plt.xlim(0, 5)
plt.ylim(0, 6)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.show()
```

# Exercício 9: Composição de Transformações

```python
# Exercício 9: Cisalhamento Vertical
def shear_vertical(x, y, k):
    return (x, y + k * x)

P_shear_vertical = (2, 3)
sheared_P_vertical = shear_vertical(*P_shear_vertical, k=1)

# Plotando o Exercício 9
plt.figure(figsize=(10, 6))
plt.title('Exercício 9: Cisalhamento Vertical')
plot_points([P_shear_vertical, sheared_P_vertical], color='blue', label='P e P\'')
plt.text(P_shear_vertical[0], P_shear_vertical[1], f'P{P_shear_vertical}', fontsize=12, ha='right')
plt.text(sheared_P_vertical[0], sheared_P_vertical[1], f"P'({sheared_P_vertical[0]}, {sheared_P_vertical[1]})", fontsize=12, ha='right')
plt.xlim(0, 5)
plt.ylim(0, 6)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.show()
```

# Exercício 10: Combinação de Transformações em uma Figura

```python
# Exercício 10: Combinação de Transformações
def transform_point(x, y):
    # Cisalhamento Vertical e Horizontal
    y = y + 2 * x  # Cisalhamento Vertical
    x = x + 2 * y  # Cisalhamento Horizontal
    return (x, y)

P_combined = (1, 1)
transformed_P = transform_point(*P_combined)

# Plotando o Exercício 10
plt.figure(figsize=(10, 6))
plt.title('Exercício 10: Combinação de Transformações')
plot_points([P_combined, transformed_P], color='blue', label='P e P\'')
plt.text(P_combined[0], P_combined[1], f'P{P_combined}', fontsize=12, ha='right')
plt.text(transformed_P[0], transformed_P[1], f"P'({transformed_P[0]}, {transformed_P[1]})", fontsize=12, ha='right')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.show()
```
