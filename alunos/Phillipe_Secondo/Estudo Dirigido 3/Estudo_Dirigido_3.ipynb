##Importação da imagem

```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import cv2

# Fazer upload da imagem
uploaded = files.upload()

# Carregar a imagem usando PIL
imagem = Image.open(list(uploaded.keys())[0])

# Converter a imagem PIL para um formato que o OpenCV entende
image = cv2.cvtColor(cv2.imread(list(uploaded.keys())[0]), cv2.COLOR_RGB2BGR)

plt.imshow(imagem)
plt.axis('off')  # Desativa os eixos
plt.show()
```

##Converter para HSV


```python
# Converter para HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt.imshow(image_hsv)
plt.axis('off')  # Desativa os eixos
plt.show()

```

##Converter para CIE LAB


```python
# Converter para CIE LAB
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)

plt.imshow(image_lab)
plt.axis('off')  # Desativa os eixos
plt.show()
```
