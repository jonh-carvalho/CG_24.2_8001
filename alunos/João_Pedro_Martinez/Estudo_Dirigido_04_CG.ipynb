```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vetor_inicial = np.array([2, 3])
vetor_translacao = np.array([4, -2])

vetor_final = vetor_inicial + vetor_translacao
print(vetor_final)

fig, ax = plt.subplots()
ax.quiver(0, 0, vetor_inicial[0], vetor_inicial[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(0 + vetor_translacao[0], 0 + vetor_translacao[1], vetor_inicial[0], vetor_inicial[1], angles='xy', scale_units='xy', scale=1, color='b')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.grid()
plt.show()
```

```python
triangulo = np.array([[1, 1],
                      [3, 1],
                      [2, 4]])

fig, ax = plt.subplots()
ax.plot((triangulo[0][0], triangulo[1][0], triangulo[2][0], triangulo[0][0]),
        (triangulo[0][1], triangulo[1][1], triangulo[2][1], triangulo[0][1]),
        color='r')

plt.xlim(0, 10)
plt.ylim(0, 10)

triangulo_escalonado = triangulo * 2

ax.plot((triangulo_escalonado[0][0], triangulo_escalonado[1][0], triangulo_escalonado[2][0], triangulo_escalonado[0][0]),
        (triangulo_escalonado[0][1], triangulo_escalonado[1][1], triangulo_escalonado[2][1], triangulo_escalonado[0][1]),
        color='b')

plt.show()
```

```python
triangulo = np.array([[1.0, 1.0],
                      [3.0, 1.0],
                      [2.0, 4.0]])

matriz_escala = np.array([[2.0, 0],
                          [0, 0.5]])

fig, ax = plt.subplots()
ax.plot((triangulo[0][0], triangulo[1][0], triangulo[2][0], triangulo[0][0]),
        (triangulo[0][1], triangulo[1][1], triangulo[2][1], triangulo[0][1]),
        color='r')

plt.xlim(0, 10)
plt.ylim(0, 10)

triangulo_escalonado = triangulo

for i in range(len(triangulo)):
    print(triangulo[i])
    triangulo_escalonado[i] = np.matmul(triangulo[i], matriz_escala)

print(triangulo_escalonado)

ax.plot((triangulo_escalonado[0][0], triangulo_escalonado[1][0], triangulo_escalonado[2][0], triangulo_escalonado[0][0]),
        (triangulo_escalonado[0][1], triangulo_escalonado[1][1], triangulo_escalonado[2][1], triangulo_escalonado[0][1]),
        color='b')

plt.show()
```

```python
triangulo = np.array([[1.0, 1.0],
                      [3.0, 1.0],
                      [2.0, 4.0]])

cos45 = np.sqrt(2)/2

matriz_rotacao = np.array([[cos45, -cos45],
                          [cos45, cos45]])

triangulo_rotacionado = np.ones(triangulo.shape)

for i in range(len(triangulo)):
    triangulo_rotacionado[i] *= np.matmul(matriz_rotacao, triangulo[i])

fig, ax = plt.subplots()
ax.plot((triangulo[0][0], triangulo[1][0], triangulo[2][0], triangulo[0][0]),
        (triangulo[0][1], triangulo[1][1], triangulo[2][1], triangulo[0][1]),
        color='r')
ax.plot((triangulo_rotacionado[0][0], triangulo_rotacionado[1][0], triangulo_rotacionado[2][0], triangulo_rotacionado[0][0]),
        (triangulo_rotacionado[0][1], triangulo_rotacionado[1][1], triangulo_rotacionado[2][1], triangulo_rotacionado[0][1]),
        color='b')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.grid()
plt.show()
```

```python
triangulo = np.array([[1.0, 1.0],
                      [3.0, 1.0],
                      [2.0, 4.0]])

matriz_reflexao = np.array([[-1, 0],
                            [0, 1]])

triangulo_refletido = np.ones(triangulo.shape)

for i in range(len(triangulo)):
    triangulo_refletido[i] *= np.matmul(matriz_reflexao, triangulo[i])

fig, ax = plt.subplots()
ax.plot((triangulo[0][0], triangulo[1][0], triangulo[2][0], triangulo[0][0]),
        (triangulo[0][1], triangulo[1][1], triangulo[2][1], triangulo[0][1]),
        color='r')
ax.plot((triangulo_refletido[0][0], triangulo_refletido[1][0], triangulo_refletido[2][0], triangulo_refletido[0][0]),
        (triangulo_refletido[0][1], triangulo_refletido[1][1], triangulo_refletido[2][1], triangulo_refletido[0][1]),
        color='b')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.grid()
plt.show()
```

```python
quadrado = np.array([[.0, .0],
                     [1.0, .0],
                     [1.0, 1.0],
                     [.0, 1.0]])

matriz_cisalhamento = np.array([[.10, 2.0],
                                [.0, 1.0]])

quadrado_cisalhado = np.ones(quadrado.shape)

for i in range(len(quadrado)):
    quadrado_cisalhado[i] *= np.matmul(matriz_cisalhamento, quadrado[i])

fig, ax = plt.subplots()
ax.plot((quadrado[0][0], quadrado[1][0], quadrado[2][0], quadrado[3][0], quadrado[0][0]),
        (quadrado[0][1], quadrado[1][1], quadrado[2][1], quadrado[3][1], quadrado[0][1]),
        color='r')
ax.plot((quadrado_cisalhado[0][0], quadrado_cisalhado[1][0], quadrado_cisalhado[2][0], quadrado_cisalhado[3][0], quadrado_cisalhado[0][0]),
        (quadrado_cisalhado[0][1], quadrado_cisalhado[1][1], quadrado_cisalhado[2][1], quadrado_cisalhado[3][1], quadrado_cisalhado[0][1]),
        color='b')

plt.xlim(0, 5)
plt.ylim(0, 5)

plt.grid()
plt.show()
```

```python
vetor_inicial = np.array([3, 2])

ponto_zero = np.array([0, 0])

vetor_translacao = np.array([1, -1])

matriz_rotacao = np.array([[0, -1],
                           [1, 0]])

fig, ax = plt.subplots()
ax.quiver(0, 0, vetor_inicial[0], vetor_inicial[1], angles='xy', scale_units='xy', scale=1, color='r')

vetor_final = (vetor_inicial + vetor_translacao)
ponto_zero += vetor_translacao

ax.quiver(ponto_zero[0], ponto_zero[1], vetor_inicial[0], vetor_inicial[1], angles='xy', scale_units='xy', scale=1, color='b')

vetor_final = np.matmul(matriz_rotacao, vetor_final)

ax.quiver(ponto_zero[0], ponto_zero[1], vetor_final[0], vetor_final[1], angles='xy', scale_units='xy', scale=1, color='g')

vetor_final *= 2

ax.quiver(ponto_zero[0], ponto_zero[1], vetor_final[0], vetor_final[1], angles='xy', scale_units='xy', scale=1, color='y')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.legend(['Original', 'Transladado', 'Transladado+Rotacionado', 'Transladado+Rotacionado+Escalonado'], loc='lower left')

plt.grid()
plt.show()

print(f'Ponto Inicial = P({vetor_inicial[0]}, {vetor_inicial[1]})')
print(f'Ponto Final = P({vetor_final[0]}, {vetor_final[1]})')
```

```python
retangulo = np.array([[1.0, 1.0],
                      [5.0, 1.0],
                      [5.0, 3.0],
                      [1.0, 3.0]])

matriz_translacao = np.array([-2.0, 3.0])

matriz_escalonamento = np.array([[1.5, 0.0],
                                 [0.0, 0.5]])

matriz_reflexao = np.array([[-1, 0],
                            [0, 1]])

retangulo_final = np.ones(retangulo.shape)

for i in range(len(retangulo)):
    retangulo_final[i] = [retangulo[i][0] + matriz_translacao[0],
                          retangulo[i][1] + matriz_translacao[1]]

fig, ax = plt.subplots()
ax.plot((retangulo[0][0], retangulo[1][0], retangulo[2][0], retangulo[3][0], retangulo[0][0]),
        (retangulo[0][1], retangulo[1][1], retangulo[2][1], retangulo[3][1], retangulo[0][1]),
        color='r')
ax.plot((retangulo_final[0][0], retangulo_final[1][0], retangulo_final[2][0], retangulo_final[3][0], retangulo_final[0][0]),
        (retangulo_final[0][1], retangulo_final[1][1], retangulo_final[2][1], retangulo_final[3][1], retangulo_final[0][1]),
        color='b')

for i in range(len(retangulo_final)):
    retangulo_final[i] = np.matmul(matriz_escalonamento, retangulo_final[i])

ax.plot((retangulo_final[0][0], retangulo_final[1][0], retangulo_final[2][0], retangulo_final[3][0], retangulo_final[0][0]),
        (retangulo_final[0][1], retangulo_final[1][1], retangulo_final[2][1], retangulo_final[3][1], retangulo_final[0][1]),
        color='g')

for i in range(len(retangulo_final)):
    retangulo_final[i] = np.matmul(matriz_reflexao, retangulo_final[i])

ax.plot((retangulo_final[0][0], retangulo_final[1][0], retangulo_final[2][0], retangulo_final[3][0], retangulo_final[0][0]),
        (retangulo_final[0][1], retangulo_final[1][1], retangulo_final[2][1], retangulo_final[3][1], retangulo_final[0][1]),
        color='y')

plt.xlim(-8, 8)
plt.ylim(-3, 9)

plt.legend(['Original', 'Transladado', 'Transladado+Escalonado', 'Transladado+Escalonado+Refletido'], loc='lower left')

plt.grid()
plt.show()
```
