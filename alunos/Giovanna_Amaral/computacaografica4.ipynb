```python
import numpy as np
import matplotlib.pyplot as plt

# Fun√ß√µes de Transforma√ß√£o
def translate(point, vector):
    return point + vector

def scale(points, factor):
    return points * factor

def scale_non_uniform(points, scale_x, scale_y):
    return points * [scale_x, scale_y]

def rotate(point, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([[np.cos(angle_radians), -np.sin(angle_radians)],
                                [np.sin(angle_radians),  np.cos(angle_radians)]])
    return point @ rotation_matrix

def reflect_y_axis(point):
    return point * [-1, 1]

def reflect_x_axis(points):
    return points * [1, -1]

def shear_horizontal(point, k):
    shear_matrix = np.array([[1, k],
                             [0, 1]])
    return point @ shear_matrix

def plot_points(original, transformed, title, labels=("Original", "Transformado")):
    plt.figure()
    plt.scatter(*zip(*original), label=labels[0], color='blue')
    plt.scatter(*zip(*transformed), label=labels[1], color='red')
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid()
    plt.show()

# Exerc√≠cio 1: Transla√ß√£o Simples
P = np.array([2, 3])
translation_vector = np.array([4, -2])
P_prime = translate(P, translation_vector)
print("Exerc√≠cio 1 - Novo ponto P':", P_prime)
plot_points([P], [P_prime], "Exerc√≠cio 1: Transla√ß√£o de um Ponto")

# Exerc√≠cio 2: Escala Uniforme em um Tri√¢ngulo
triangle = np.array([[1, 1], [3, 1], [2, 4]])
scaled_triangle = scale(triangle, 2)
print("Exerc√≠cio 2 - V√©rtices ap√≥s escala uniforme:", scaled_triangle)
plot_points(triangle, scaled_triangle, "Exerc√≠cio 2: Escala Uniforme em um Tri√¢ngulo")

# Exerc√≠cio 3: Escala N√£o Uniforme
non_uniform_scaled_triangle = scale_non_uniform(triangle, 2, 0.5)
print("Exerc√≠cio 3 - V√©rtices ap√≥s escala n√£o uniforme:", non_uniform_scaled_triangle)
plot_points(triangle, non_uniform_scaled_triangle, "Exerc√≠cio 3: Escala N√£o Uniforme")

# Exerc√≠cio 4: Rota√ß√£o em Torno da Origem
P = np.array([1, 0])
P_rotated = rotate(P, 90)
print("Exerc√≠cio 4 - Novo ponto P' ap√≥s rota√ß√£o de 90¬∞:", P_rotated)
plot_points([P], [P_rotated], "Exerc√≠cio 4: Rota√ß√£o de um Ponto")

# Exerc√≠cio 5: Rota√ß√£o de um Quadrado
square = np.array([[1, 1], [1, 4], [4, 4], [4, 1]])
rotated_square = np.array([rotate(point, -45) for point in square])
print("Exerc√≠cio 5 - V√©rtices ap√≥s rota√ß√£o de 45¬∞:", rotated_square)
plot_points(square, rotated_square, "Exerc√≠cio 5: Rota√ß√£o de um Quadrado")

# Exerc√≠cio 6: Reflex√£o Simples em rela√ß√£o ao eixo y
P = np.array([2, 5])
P_reflected = reflect_y_axis(P)
print("Exerc√≠cio 6 - Novo ponto P' ap√≥s reflex√£o:", P_reflected)
plot_points([P], [P_reflected], "Exerc√≠cio 6: Reflex√£o de um Ponto")

# Exerc√≠cio 7: Reflex√£o de um Tri√¢ngulo em rela√ß√£o ao eixo x
triangle = np.array([[2, 3], [4, 3], [3, 5]])
reflected_triangle = reflect_x_axis(triangle)
print("Exerc√≠cio 7 - V√©rtices ap√≥s reflex√£o:", reflected_triangle)
plot_points(triangle, reflected_triangle, "Exerc√≠cio 7: Reflex√£o de um Tri√¢ngulo")

# Exerc√≠cio 8: Cisalhamento Horizontal
P = np.array([2, 3])
P_sheared = shear_horizontal(P, 2)
print("Exerc√≠cio 8 - Novo ponto P' ap√≥s cisalhamento:", P_sheared)
plot_points([P], [P_sheared], "Exerc√≠cio 8: Cisalhamento Horizontal")

# Exerc√≠cio 9: Composi√ß√£o de Transforma√ß√µes
P = np.array([3, 2])
P_translated = translate(P, [1, -1])
P_rotated = rotate(P_translated, 90)
P_scaled = scale(P_rotated, 2)
print("Exerc√≠cio 9 - Novo ponto P' ap√≥s composi√ß√£o de transforma√ß√µes:", P_scaled)
plot_points([P], [P_scaled], "Exerc√≠cio 9: Composi√ß√£o de Transforma√ß√µes")

# Exerc√≠cio 10: Combina√ß√£o de Transforma√ß√µes em um Ret√¢ngulo
rectangle = np.array([[1, 1], [5, 1], [5, 3], [1, 3]])
# Transla√ß√£o
rectangle_translated = translate(rectangle, [-2, 3])
# Escala N√£o Uniforme
rectangle_scaled = scale_non_uniform(rectangle_translated, 1.5, 0.5)
# Reflex√£o em rela√ß√£o ao eixo y
rectangle_reflected = reflect_y_axis(rectangle_scaled)
print("Exerc√≠cio 10 - V√©rtices ap√≥s as transforma√ß√µes:", rectangle_reflected)
plot_points(rectangle, rectangle_reflected, "Exerc√≠cio 10: Combina√ß√£o de Transforma√ß√µes")
```
