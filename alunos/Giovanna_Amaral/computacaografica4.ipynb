```python
import numpy as np
import matplotlib.pyplot as plt

# Funções de Transformação
def translate(point, vector):
    return point + vector

def scale(points, factor):
    return points * factor

def scale_non_uniform(points, scale_x, scale_y):
    return points * [scale_x, scale_y]

def rotate(point, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([[np.cos(angle_radians), -np.sin(angle_radians)],
                                [np.sin(angle_radians),  np.cos(angle_radians)]])
    return point @ rotation_matrix

def reflect_y_axis(point):
    return point * [-1, 1]

def reflect_x_axis(points):
    return points * [1, -1]

def shear_horizontal(point, k):
    shear_matrix = np.array([[1, k],
                             [0, 1]])
    return point @ shear_matrix

def plot_points(original, transformed, title, labels=("Original", "Transformado")):
    plt.figure()
    plt.scatter(*zip(*original), label=labels[0], color='blue')
    plt.scatter(*zip(*transformed), label=labels[1], color='red')
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid()
    plt.show()

# Exercício 1: Translação Simples
P = np.array([2, 3])
translation_vector = np.array([4, -2])
P_prime = translate(P, translation_vector)
print("Exercício 1 - Novo ponto P':", P_prime)
plot_points([P], [P_prime], "Exercício 1: Translação de um Ponto")

# Exercício 2: Escala Uniforme em um Triângulo
triangle = np.array([[1, 1], [3, 1], [2, 4]])
scaled_triangle = scale(triangle, 2)
print("Exercício 2 - Vértices após escala uniforme:", scaled_triangle)
plot_points(triangle, scaled_triangle, "Exercício 2: Escala Uniforme em um Triângulo")

# Exercício 3: Escala Não Uniforme
non_uniform_scaled_triangle = scale_non_uniform(triangle, 2, 0.5)
print("Exercício 3 - Vértices após escala não uniforme:", non_uniform_scaled_triangle)
plot_points(triangle, non_uniform_scaled_triangle, "Exercício 3: Escala Não Uniforme")

# Exercício 4: Rotação em Torno da Origem
P = np.array([1, 0])
P_rotated = rotate(P, 90)
print("Exercício 4 - Novo ponto P' após rotação de 90°:", P_rotated)
plot_points([P], [P_rotated], "Exercício 4: Rotação de um Ponto")

# Exercício 5: Rotação de um Quadrado
square = np.array([[1, 1], [1, 4], [4, 4], [4, 1]])
rotated_square = np.array([rotate(point, -45) for point in square])
print("Exercício 5 - Vértices após rotação de 45°:", rotated_square)
plot_points(square, rotated_square, "Exercício 5: Rotação de um Quadrado")

# Exercício 6: Reflexão Simples em relação ao eixo y
P = np.array([2, 5])
P_reflected = reflect_y_axis(P)
print("Exercício 6 - Novo ponto P' após reflexão:", P_reflected)
plot_points([P], [P_reflected], "Exercício 6: Reflexão de um Ponto")

# Exercício 7: Reflexão de um Triângulo em relação ao eixo x
triangle = np.array([[2, 3], [4, 3], [3, 5]])
reflected_triangle = reflect_x_axis(triangle)
print("Exercício 7 - Vértices após reflexão:", reflected_triangle)
plot_points(triangle, reflected_triangle, "Exercício 7: Reflexão de um Triângulo")

# Exercício 8: Cisalhamento Horizontal
P = np.array([2, 3])
P_sheared = shear_horizontal(P, 2)
print("Exercício 8 - Novo ponto P' após cisalhamento:", P_sheared)
plot_points([P], [P_sheared], "Exercício 8: Cisalhamento Horizontal")

# Exercício 9: Composição de Transformações
P = np.array([3, 2])
P_translated = translate(P, [1, -1])
P_rotated = rotate(P_translated, 90)
P_scaled = scale(P_rotated, 2)
print("Exercício 9 - Novo ponto P' após composição de transformações:", P_scaled)
plot_points([P], [P_scaled], "Exercício 9: Composição de Transformações")

# Exercício 10: Combinação de Transformações em um Retângulo
rectangle = np.array([[1, 1], [5, 1], [5, 3], [1, 3]])
# Translação
rectangle_translated = translate(rectangle, [-2, 3])
# Escala Não Uniforme
rectangle_scaled = scale_non_uniform(rectangle_translated, 1.5, 0.5)
# Reflexão em relação ao eixo y
rectangle_reflected = reflect_y_axis(rectangle_scaled)
print("Exercício 10 - Vértices após as transformações:", rectangle_reflected)
plot_points(rectangle, rectangle_reflected, "Exercício 10: Combinação de Transformações")
```
