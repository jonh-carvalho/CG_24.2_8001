```python
from PIL import Image
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML

# 1. Redimensionamento e Cropping
image_path = 'imagem.jpeg'
img = Image.open(image_path)

# Redimensionar a imagem para diferentes tamanhos
sizes = [(200, 200), (400, 400), (800, 800)]
resized_images = [img.resize(size) for size in sizes]

# Fazer um recorte (cropping) da área central da imagem
width, height = img.size
left = (width - 200) / 2
top = (height - 200) / 2
right = (width + 200) / 2
bottom = (height + 200) / 2

cropped_image = img.crop((left, top, right, bottom))

# Salvar as imagens processadas
for i, resized_img in enumerate(resized_images):
    resized_img.save(f'resized_image_{sizes[i][0]}x{sizes[i][1]}.jpg')

cropped_image.save('cropped_image.jpg')

# Mostrar as imagens redimensionadas e a imagem cortada
plt.figure(figsize=(12, 6))
for i, resized_img in enumerate(resized_images):
    plt.subplot(2, len(sizes), i + 1)
    plt.imshow(resized_img)
    plt.axis('off')
    plt.title(f'Resized {sizes[i]}')

plt.subplot(2, len(sizes), len(sizes) + 1)
plt.imshow(cropped_image)
plt.axis('off')
plt.title('Cropped Image')

plt.show()

# 2. Conversão de Cores
def apply_sepia(image):
    sepia_filter = Image.new('RGB', image.size)
    pixels = image.load()
    for y in range(image.height):
        for x in range(image.width):
            r, g, b = pixels[x, y]
            tr = int(0.393 * r + 0.769 * g + 0.189 * b)
            tg = int(0.349 * r + 0.686 * g + 0.168 * b)
            tb = int(0.272 * r + 0.534 * g + 0.131 * b)
            sepia_filter.putpixel((x, y), (min(tr, 255), min(tg, 255), min(tb, 255)))
    return sepia_filter

color_image_path = 'imagem.jpeg'
color_img = Image.open(color_image_path)

# Converter para escala de cinza
gray_image = color_img.convert('L')

# Aplicar filtro sepia
sepia_image = apply_sepia(color_img)

# Salvar as imagens processadas
gray_image.save('gray_image.jpg')
sepia_image.save('sepia_image.jpg')

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(color_img)
plt.axis('off')
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Gray')

plt.subplot(1, 3, 3)
plt.imshow(sepia_image)
plt.axis('off')
plt.title('Sepia')

plt.show()


```
